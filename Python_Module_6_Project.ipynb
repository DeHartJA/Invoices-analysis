In order to answer the question of which customer has the most invoices , we first need to group by customer and invoiceID, by using the nunique function we can check for each unique invoice from that customer,next we find the max number of those invoices and which customer had the most unique invoices. in this case the answer is Stuff by Stew with 31,198 invoices.

```python

# Count the number of unique invoices for each customer
unique_invoices_per_customer = df.groupby('Customer')['InvoiceID'].nunique()

# Determine the customer with the most unique invoices
most_unique_invoices_customer = unique_invoices_per_customer.idxmax()
num_unique_invoices = unique_invoices_per_customer.max()

print("Customer with the most unique invoices:", most_unique_invoices_customer)
print("Number of unique invoices:", num_unique_invoices)
```

When we want to fill in missing data we first have to identify ways that can be done, in this case we can use the StockItemID and Description, if we know what both of these are then we can fill in description if we know what the stockitemid is and vice versa

Question 8

```python
# drop rows in the StockItemID and Description columns if either column have null values.
df.dropna(subset=['StockItemID', 'Description'])
```

```python
# find the value counts of the StockItemID column
df['StockItemID'].value_counts()
```

The code below matches up StockItemID and Description to Impute the null values

```python
# define the imput description function, if StockItemID is not null and description is null, fill in the description

# Function to impute Description based on StockItemID
def impute_description(row):
    if pd.notnull(row['StockItemID']):
        description = df.loc[df['StockItemID'] == row['StockItemID'], 'Description'].values
        if len(description) > 0:
            return description[0]
    return None

# Apply the function to impute Description
df['Description'] = df.apply(impute_description, axis=1)

print(df)
```

After applying the above code we can drop the rows for StockItemID and Description if both are Null

```python
# drop rows that show null values in both the StockItemID and the Description column
df.dropna(subset=['StockItemID', 'Description'])
```

This code multiplies the unit price by quantity and taxrate and then divids it by 100 to get the TaxAmount, this code also fills in any null values in the TaxAmount column so there are no null values in that column now.

```python
df['TaxAmount'] = df['TaxAmount'].fillna((df['UnitPrice'] * df['Quantity'] * df['TaxRate'] / 100))
```

This code multiplies the unitprice by quantity and then adds the taxamount to create the extendedprice, also fill in any null values

```python
df['ExtendedPrice'] = df['ExtendedPrice'].fillna((df['UnitPrice'] * df['Quantity']) + df['TaxAmount'])
```

This code creates a new column for zscore for the quantity and extended price. zscore is use to identify outliers in the data. Anything with a zscore of greater then 3 is considered an outlier.

```python
import pandas as pd
from scipy.stats import zscore

# Calculate z-scores for 'Quantity' and 'ExtendedPrice'
df['Quantity_zscore'] = zscore(df['Quantity'])
df['ExtendedPrice_zscore'] = zscore(df['ExtendedPrice'])

# Identify outliers based on z-score threshold of 3
outliers_quantity = df[(df['Quantity_zscore'] > 3) | (df['Quantity_zscore'] < -3)]
outliers_extendedprice = df[(df['ExtendedPrice_zscore'] > 3) | (df['ExtendedPrice_zscore'] < -3)]

print("Outliers in Quantity column:")
print(outliers_extendedprice)

```
